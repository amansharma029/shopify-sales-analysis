# Shopify Sales Analysis #

## 1. Introduction

The Shopify Sales Dataset contains transactional data from an e-commerce platform, including information about orders, products, customers, pricing, discounts, shipping costs, ratings, and returns.

 ## 2. Import libraries

```python
import pandas as pd
from sqlalchemy import create_engine
import os
from dotenv import load_dotenv
```

## 3. Load Dataset

```python
df = pd.read_csv("shopify_sales_dataset.csv")
df.head()
```

## 6. Exploratory Data Analysis

### A. Overall Business Performance

```python
total_revenue = df['revenue'].sum().round(2)
total_profit = df['profit'].sum().round(2)

total_revenue, total_profit
```

### B. Profitability Check

```python
df['Profit Margin'] = df['profit']/df['revenue']
df['Profit Margin'].describe()
```

### C. Top Customers

```python
top_customers = df.groupby('customer_id')['revenue'].sum().sort_values(ascending = False)
top_customers.head(5)
```

 Top customers contribute significant portion of revenue

### D. Product Performance 

```python
df.groupby('product_id')['revenue'].sum().sort_values(ascending = False).head(5)
```

Few products dominate revenue generation

### E. Discount Analysis

```python
df.groupby('discount_percent')['revenue'].mean()
```

Higher discount decrease revenue

### F. Rating Analysis

```python
df.groupby('rating')['revenue'].mean().sort_values(ascending = False).head(5)  
```

### G. Shipping Cost Impact

```python
df[['shipping_cost', 'revenue', 'profit']].corr()  
```
